```python
def determine_command_type(command):
	type_string = ""
	if (command == "DIRECTIVE_data" or command == "DIRECTIVE_text" or command == "DIRECTIVE_globl" or command == "DIRECTIVE_global" or command == "DIRECTIVE_end"):
		type_string = "Property"
	if (command == "DIRECTIVE_byte" or command == "DIRECTIVE_word" or command == "DIRECTIVE_long" or command == "DIRECTIVE_quad" or command == "DIRECTIVE_int" or command == "DIRECTIVE_ascii" or command == "DIRECTIVE_asciz" or command == "DIRECTIVE_align"):
		type_string = "Enum"
	if (command[0:3] == "add" or command[0:3] == "sub" or command[0:3] == "neg" or command[0:3] == "inc" or command[0:3] == "dec" or command[0:3] == "mul" or command[0:3] == "div" or command[0:4] == "imul" or command[0:4] == "idiv"):
		type_string = "Operator"
	if (command[0:3] == "shr" or command[0:3] == "shl" or command[0:3] == "sar" or command[0:3] == "sal" or command[0:3] == "rol" or command[0:3] == "ror" or command[0:3] == "rcl" or command[0:3] == "rcr"):
		type_string = "Operator"
	if (command[0:3] == "not" or command[0:3] == "and" or command[0:2] == "or" or command[0:3] == "xor"):
		type_string = "Operator"
	if (command[0:3] == "cmp" or command[0:4] == "test"):
		type_string = "Function"
	if (command[0:3] == "mov" or command[0:3] == "lea"):
		type_string = "Function"
	if (command[0:3] == "pop" or command[0:4] == "push" or command == "enter" or command == "leave" or command == "ret" or command == "loop" or command[0] == "j" or command == "call" or command == "nop" or command == "xchg" or command == "int"):
		type_string = "Event"
	if (command[-1] == "x" or command == "al" or command == "ar" or command == "bl" or command == "br" or command == "cl" or command == "cr" or command == "dl" or command == "dr" or command == "edi" or command == "di" or command == "esi" or command == "si" or command == "eip" or command == "esp" or command == "ebp"):
		type_string = "Variable"
	return type_string
```
